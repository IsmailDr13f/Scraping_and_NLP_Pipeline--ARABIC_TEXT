### 3. Establishment of NLP Pipeline (Text Cleaning, Tokenization, Stop words, Discretization, Normalization).

#### 1-Text Cleaning:

For this task we using:<br>
-Removes all diacritics(علامات التشكيل).

```python
import re
text = 'اللَّهمَّ اغْفِرْ لنَا ولوالدِينَا '    
output=re.sub(u'[\u064e\u064f\u0650\u0651\u0652\u064c\u064b\u064d\u0640\ufc62]','',text)
print(output)
#اللهم اغفر لنا ولوالدينا 
```

```python
import re
def remove_diacritics(txt):
    return re.sub(u'[\u064e\u064f\u0650\u0651\u0652\u064c\u064b\u064d\u0640\ufc62]','',txt)
```

```python
data_diacritics=[remove_diacritics(tt) for tt in data_lst]
data_diacritics
```

```python
# Function to remove special characters
def remove_special_characters(text):
    # replace \xa0 with a space
    text = text.replace('\xa0', ' ')
    # remove arabic and normale numbers
    text = re.sub(r'[٠-٩]', '', text)
    text = re.sub(r'[0-9]', '', text)
    # Remove newline character
    text = text.replace('\n', ' ')
    # remove other special characters
    return re.sub(r'[^\w\s]', '  ', text)
    #return text
```

```python
data_rmv_sc = [remove_special_characters(txt) for txt in data_diacritics]
print(data_rmv_sc[0])
```

#### 2-Tokenization:

```python
#!python -m spacy download en_core_web_sm
```

```python
# Tokenizing using sentences-tokenization:
import spacy
# Load the Arabic language model for spaCy
nlp = spacy.load('en_core_web_sm')
# Tokenize each paragraph into sentences
sentences = [list(nlp(paragraph).sents) for paragraph in data_rmv_sc]
print(len(sentences[0]))
```

```python
sentences[0]
```

```python
# Tokenize each sentence into phrase 
phrases = [[token.text for token in sentence] for sentence in sentences ]
print(len(phrases))
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize("Let's see how it's working.")

# Tokenize each phrase into words 
words = [tokenizer.tokenize(elem) for phrase in phrases for elem in phrase ]
print(len(words))
```

### 4. Do the Stemming and Lemmatization, then compare the tow mechanisms.

```python
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

#### 1- Stemming:

```python
def stemming__(vect):
    stm = SnowballStemmer('arabic')
    stemmed_wrds = [stm.stem(wrd) for wrd in vect]
    return stemmed_wrds
```

```python
stemming_lista=[stemming__(vct) for vct in words]
print(stemming_lista[0])
```

#### 2- Lemmatization:

```python
def lemmatizing__(vect):
    lem = WordNetLemmatizer()
    lemmatized_wrd = [lem.lemmatize(wrd) for wrd in vect]
    return lemmatized_wrd
```

```python
lemmatizing_lista=[lemmatizing__(vct) for vct in words]
print(lemmatizing_lista[0])
```

Note: By comparing the two methods, it's clear that the lemmatization is more performent that the stemming, because in arabic, the stricture of the words not always subject to the same linguistic rules; for example:<br> 'عل' and 'عليه'<br>'صلا' and 'الصلاة'....
